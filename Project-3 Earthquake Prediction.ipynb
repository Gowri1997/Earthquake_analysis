### Feature Engineering
We need to remove data those depth and magnitude and horizontal distance  errors are beyond than its upper and lower limit.Azimuth Gap greater than 180 will cause the location accuracy degrades significantly

```python
quake=quake[quake['Azimuthal Gap']<180]
quake
```

```python
quake.isnull().sum()
```

```python
quake.Status.unique()
```

```python
quake_1=quake
```

```python
quake_1['Longitude_r']=quake_1['Longitude'].round()
```

```python
quake_1['Latitude_r']=quake_1['Latitude'].round()
```

```python
quake_1=quake_1.drop(['ID','Source','Location Source','Magnitude Source','Latitude','Longitude','Status'],axis='columns')
```

```python
quake_1
```

```python
quake_1.info()
```

```python
quake_1=quake_1[~(quake_1['Magnitude Type']==0)]
quake_1
```

```python
quake_1.Magnitude=quake_1.Magnitude.round(1)
quake_1.groupby(by='Magnitude').count()
```

```python
plt.hist(quake_1.Magnitude,bins=20,rwidth=.8)
plt.xlabel('Magnitude')
plt.ylabel('frequency')
```

As magnitude increases the freaquency of earthquake decreases

```python
quake_1.Depth=quake_1.Depth.round(0)
quake_1.groupby(by='Depth').count()
plt.hist(quake_1.Depth,bins=20,rwidth=.8)
plt.xlabel('Depth')
plt.ylabel('frequency')
```

```python
quake_1.info()
```

```python
quake_1['Date']=pd.to_datetime(quake_1['Date'],infer_datetime_format=True,errors='coerce')
quake_1['Year']=quake_1['Date'].dt.year
quake_1['Month']=quake_1['Date'].dt.month
```

```python
quake_1.info()
```

```python
quake_1
```

```python
quake_1.Year=quake_1.Year.round(0)
quake_1.groupby(by='Year').count()
plt.hist(quake_1.Year,bins=20,rwidth=.8)
plt.xlabel('Year')
plt.ylabel('frequency')
```

Indian main land extends between latitudes 8°4'N and 37°6'N and longitudes 68°7'E and 97°25'E.so considering earthquake near India

```python
quake_india=quake_1[(quake_1.Longitude_r>68)&(quake_1.Longitude_r<98)&(quake_1.Latitude_r>8)&(quake_1.Latitude_r<38)]
quake_india
```

```python
quake_india.sort_values(by=['Year','Month'],ascending=True)
```

```python
quake_india['Time_line']=quake_india.Year+(quake_india.Month)/12
quake_india
```

```python
plt.plot(quake_india.Time_line,quake_india.Magnitude)
plt.xlabel('Time_line')
plt.ylabel('Magnitude')
```

```python
year_list=quake_india.Year.unique()
```

```python
quake_india['Interval']=np.append([0],np.diff(quake_india.Time_line))
quake_india
```
